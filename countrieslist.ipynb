### ## Scraping Wikipedia countries list with requests and bs4

```python
# Second portion

import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://simple.m.wikipedia.org/wiki/List_of_countries")
```

```python
# Create soup and get title

soup = BeautifulSoup(page.text, 'html.parser')
print(soup.title)
```

```python
# Find all list items (<li>) inside the country list section
countries_list = []
for li in soup.select("ul > li a"):
    country = li.get_text()
    countries_list.append(country)

# Print the extracted countries
print(countries_list)
```

```python
import requests
import csv
import re  # Import regular expressions module
from bs4 import BeautifulSoup

# Wikipedia URL
url = "https://simple.m.wikipedia.org/wiki/List_of_countries"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Regular expression to allow only English alphabet (A-Z, a-z, spaces, and hyphens)
english_only = re.compile(r"^[A-Za-z\s\-]+$")

# Extract country names, filtering out non-English names
countries_list = [li.get_text() for li in soup.select("ul > li a") if english_only.match(li.get_text())]

# Save to CSV file
csv_filename = "countries_list.csv"
with open(csv_filename, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Country"])  # Header
    for country in countries_list:
        writer.writerow([country])

# Print the extracted country names
print(countries_list)
print(f"Saved {len(countries_list)} countries to {csv_filename}")
```
