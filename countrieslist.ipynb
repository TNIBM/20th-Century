### ## Scraping Wikipedia countries list with requests and bs4

```python
# Second portion

import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://simple.m.wikipedia.org/wiki/List_of_countries")
```

```python
# Create soup and get title

soup = BeautifulSoup(page.text, 'html.parser')
print(soup.title)
```

```python
print(soup.text)
```

```python
text = soup.get_text()
```

```python
text = text.encode('utf-8')
```

```python
with open('countries list.txt', 'wb') as f: 
    f.write(text)
```
