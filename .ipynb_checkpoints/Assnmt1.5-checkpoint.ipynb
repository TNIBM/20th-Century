```python
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

## Tokenization

```python
myfile = open('20th_century.txt', encoding='utf-8') 
```

```python
# Import txt file

with open('20th_century.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
# Sentence tokenization 

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10,cumulative = False)
plt.show()
```

## Stopwords

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in sentences

# filtered_sent = []
# for word in tokenized_sent:
#     if word not in stop_words:
#         filtered_sent.append(word)
# print("Tokenized Sentence:", tokenized_sent)
# print("Filterd Sentence:", filtered_sent)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

```python
# Create a new FreqDist for filtered_words

dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative = False)
plt.show()
```

```python
dist_words_filter
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
sans_punc
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

### Finally after getting rid of unnecessary punctuations and stop words, we can reach to the real word data with this skimmed text.

```python
dist_words_filter_2.most_common(20)
```

## POS Tags with TextBlob

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The']
```

```python
filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
text
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
```

```python
df_text.head()
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
df_t.head()
```

### Top 10 POS tags

```python
top10 = df_t.nlargest(10, 'Words')
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top10).set_title("20th_century - top 10 word types used")
```

### Nouns

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
top15
```

```python
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15).set_title("20th_century - most frequently used nouns")
```

### Verbs

```python
df = df_text[(df_text['Word type'] == "VB")  | (df_text['Word type'] == "VBD")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
top15
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x = "Word", y = "Occurences",
    saturation = 0.9, data = top15).set_title("20th_century - most frequently used verbs")
```

### Adjectives

```python
df = df_text[df_text['Word type'] == "JJ"]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by=['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
top15
```

```python
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data=top15).set_title("20th_century - most frequently used adjectives")
```

## Number of times countries are mentioned in the text

```python
listToStr = ' '.join([str(elem) for elem in filtered])
 
print(listToStr)
```

```python
# Create a count for the countries

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
all_counts
```

```python
country_name = pd.read_csv("countries_list_20th_century_1.5.csv", index_col = 0)
```

```python
country_name.head()
```

```python
country_name
```

```python
# Convert countries list df to list
country_name = country_name['country_name'].str.strip().tolist()
```

```python
# Create a list to hold matches
filtered_countries = []

# Iterate through the tokenized words
for i in range(len(filtered)):
    # Check single-word matches
    if filtered[i] in country_name:
        filtered_countries.append(filtered[i])
    
    # Check two-word sequences
    if i < len(filtered) - 1:
        two_word = f"{filtered[i]} {filtered[i + 1]}"
        if two_word in country_name:
            filtered_countries.append(two_word)
```

```python
# Count frequency
country_counts = Counter(filtered_countries)
```

```python
print(country_counts)
```

```python
# Create Dataframe
df_country = pd.DataFrame.from_dict(country_counts, orient='index', columns=['Frequency']).reset_index()
```

```python
# Rename index col
df_country = df_country.rename(columns={'index':'Country'})
```

```python
# Oder Country Frequency
country_ord = df_country.sort_values(by='Frequency', ascending=False).head(10)
```

```python
# Palette
palette = sns.color_palette('light:#03719c_r', as_cmap=False, n_colors=10)
```

```python
# Plot frequency of countries
plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Frequency', data=country_ord, palette=palette)
plt.title('Top 10 Countries by Mention')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()
```

The nations most frequently mentioned are predominantly associated with wars, specifically the two World Wars and the Vietnam War. The three countries that appear most often in the Wikipedia article are Germany, Japan, and the United States.
